```python
saved_tracks = sp.current_user_saved_tracks(limit=50)

track_ids = [track['track']['id'] for track in saved_tracks['items']]
audio_features = sp.audio_features(track_ids)

# print(json.dumps(track_ids, sort_keys=True, indent=4))
danceability = [track['danceability'] for track in audio_features]
energy = [track['energy'] for track in audio_features]
valence = [track['valence'] for track in audio_features]

# Plot the distributions
plt.figure(figsize=(10, 10))
plt.hist(danceability, bins=20, alpha=0.5, label='Danceability')
plt.hist(energy, bins=20, alpha=0.5, label='Energy')
plt.hist(valence, bins=20, alpha=0.5, label='Valence')
plt.xlabel('Value')
plt.ylabel('Count')
plt.title('Audio Feature Distribution')
plt.legend()
plt.show()
```
